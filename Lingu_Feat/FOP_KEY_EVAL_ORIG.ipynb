```python
from data_util.eda import *
import os

def extract_POS(sent):
    POS_fops = FO_rule_POS(sent).run()                    
    POS_PF = []
    POS_O = []
    for f, o in POS_fops:
        POS_PF.append(f)
        POS_O.append(o)

    POS_Key = POS_PF + POS_O
    return POS_Key, POS_PF

extract_POS(sent)
```

```python
fn = 'keyword_eval/Keyword/train.xlsx'
train_df = pd.read_excel(fn, usecols=['keyword','count','sent'])
fn = 'keyword_eval/Keyword/test.xlsx'
test_df = pd.read_excel(fn, usecols=['keyword','count','sent'])
```

```python
total_df = pd.concat([train_df,test_df])
total_df = total_df.reset_index(drop=True)
total_df.head()
```

```python
total_df['label_keyword'] = total_df['keyword'].apply(crop_all)
total_df['label_count'] = total_df['label_keyword'].apply(lambda x: len(x))
```

```python
for idx in range(len(total_df)):           
    series = total_df.iloc[idx]
    label_keys, label_count, sent = series['label_keyword'], series['label_count'], series['sent']
    # pos
    pred_pos_keys = extract_POS(sent)[0]
    total_df.loc[idx,'pred_pos_keys'] = str(pred_pos_keys)
    correct_pos_keys = set(label_keys) & set(pred_pos_keys)
    total_df.loc[idx,'correct_pos_keys'] = str(correct_pos_keys)
    total_df.loc[idx,'pos_keys_count'] = len(pred_pos_keys)
    total_df.loc[idx,'correct_pos_keys_count'] = len(correct_pos_keys)

    # dep
    pred_dep_keys = extract_DEP(sent)[0]
    total_df.loc[idx,'pred_dep_keys'] = str(pred_dep_keys)
    correct_dep_keys = set(label_keys) & set(pred_dep_keys)
    total_df.loc[idx,'correct_dep_keys'] = str(correct_dep_keys)
    total_df.loc[idx,'dep_keys_count'] = len(pred_dep_keys)
    total_df.loc[idx,'correct_dep_keys_count'] = len(correct_dep_keys)
```

```python
precision = sum(list(total_df['correct_pos_keys_count']))/sum(list(total_df['pos_keys_count']))
recall = sum(list(total_df['correct_pos_keys_count']))/sum(list(total_df['label_count']))
f_measure = (2*precision*recall) / (precision + recall)
print('Keywords pos precision : %.2f'%(precision))
print('Keywords pos recall : %.2f'%(recall))
print('Keywords pos f_measure : %.2f'%(f_measure))

print('------------------------------------------------------------------------------')
precision = sum(list(total_df['correct_dep_keys_count']))/sum(list(total_df['dep_keys_count']))
recall = sum(list(total_df['correct_dep_keys_count']))/sum(list(total_df['label_count']))
f_measure = (2*precision*recall) / (precision + recall)
print('Keywords dep precision : %.2f'%(precision))
print('Keywords dep recall : %.2f'%(recall))
print('Keywords dep f_measure : %.2f'%(f_measure))
```
